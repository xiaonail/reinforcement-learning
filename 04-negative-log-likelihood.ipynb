## Negative Log-Likelihood (NLL)

The likelihood measures how probable a participant's observed sequence of choices is, given a set of parameters (α, β). 

$$\text{NLL} = -\sum_{t=1}^{T} \log P(\text{observed choice}_t \mid V^{(t)}, \alpha, \beta)$$

When the model assigns high probability to the choices the participant actually made, each term in the sum is small, producing a low NLL. So lower NLL means better fits, and better fits mean the model explains the observed behaviour better.

```python
def nll_model1(params, choices, outcomes):
    alpha, beta = params
    Va, Vb = V0, V0
    nll = 0.0
    for t in range(len(choices)):
        exp_a = np.exp(-beta * Va)
        exp_b = np.exp(-beta * Vb)
        p_a = exp_a / (exp_a + exp_b)

        # Log-prob of the observed choice
        nll -= np.log(max(p_a, 1e-12)) if choices[t] == 1 else np.log(max(1 - p_a, 1e-12))

        # Update chosen stimulus
        if choices[t] == 1:
            Va += alpha * (outcomes[t] - Va)
        else:
            Vb += alpha * (outcomes[t] - Vb)
    return nll

print("nll_model1() defined")
```

```python
# Verify with hint that subject 2 NLL ≈ 64, and report subject 1 and 10
test_params = [0.3, 8]    # note: beta = +8 here, as stated in the assignment
nll_s1  = nll_model1(test_params, choices[0],  outcomes[0])
nll_s2  = nll_model1(test_params, choices[1],  outcomes[1])
nll_s10 = nll_model1(test_params, choices[9],  outcomes[9])

print(f"NLL subject 1  (α=0.3, β=8): {nll_s1:.4f}")
print(f"NLL subject 2  (α=0.3, β=8): {nll_s2:.4f}  hint 64")
print(f"NLL subject 10 (α=0.3, β=8): {nll_s10:.4f}")
```
