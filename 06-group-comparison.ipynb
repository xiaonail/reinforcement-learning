```python
print("Modle 1 Group comparison")
for idx, name in enumerate(["α", "β"]):
    t, p = ttest_ind(fitted_m1[:25, idx], fitted_m1[25:, idx], equal_var=False)
    sig = "***" if p<0.001 else "**" if p<0.01 else "*" if p<0.05 else "n.s."
    print(f"  {name}:  t = {t:.4f},  p = {p:.4f}  {sig}")
```
