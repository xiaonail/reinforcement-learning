```python
## Model Fitting

Find the $\alpha$ and $\beta$ that minimise NLL for each participant using Nelder-Mead optimisation.
```

```python
def fit_model(nll_func, choices, outcomes, x0):
    """Fit a model by minimising NLL via Nelder-Mead."""
    return minimize(nll_func, x0, args=(choices, outcomes),
                    method="Nelder-Mead",
                    options={"maxiter": 10000, "xatol": 1e-8, "fatol": 1e-8})

# Fit Model 1 to real data for all 50 participants
x0_m1 = [0.3, -8]
fitted_m1 = np.array([fit_model(nll_model1, choices[i], outcomes[i], x0_m1).x
                       for i in range(N_PARTICIPANTS)])

alphas_m1 = fitted_m1[:, 0]
betas_m1  = fitted_m1[:, 1]

print(f"α mean: {alphas_m1.mean():.4f},  var: {alphas_m1.var():.4f}")
print(f"β mean: {betas_m1.mean():.4f},  var: {betas_m1.var():.4f}")
```

```python
# Plot to see if there is a visible difference between the high-anxiety group (indices 0–24) and the low-anxiety group (25–49). 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
colors = ["tab:red"] * 25 + ["tab:blue"] * 25

for ax, vals, lab in zip(axes, [alphas_m1, betas_m1], ["α (learning rate)", "β (inverse temp.)"]):
    ax.bar(range(N_PARTICIPANTS), vals, color=colors, alpha=0.7)
    ax.axvline(24.5, color="black", ls="--", lw=1.5)
    ax.set_xlabel("Participant"); ax.set_ylabel(lab)

axes[0].set_title("Fitted α (Model 1)"); axes[1].set_title("Fitted β (Model 1)")
plt.tight_layout(); plt.savefig("figures/task_e.png"); plt.show()
```
