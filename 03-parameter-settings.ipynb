```python
## Exploring the Parameter Settings

Vary alpha and beta systematically and observe how the expected number of aversive stimuli changes.
```

```python
# vary α and β
alphas_grid = np.linspace(0.01, 0.99, 20)
betas_grid  = np.linspace(-10, -0.1, 20)
n_sims_cell = 100
results = np.zeros((len(alphas_grid), len(betas_grid)))

print("Running parameter grid")
for i, a in enumerate(alphas_grid):
    for j, b in enumerate(betas_grid):
        results[i, j] = np.mean([simulate_model1(a, b)[1].sum() for _ in range(n_sims_cell)])
print("Done")
```

```python
# plot a heatmap
fig, ax = plt.subplots(figsize=(9, 7))
im = ax.imshow(results, origin="lower", aspect="auto",
               extent=[betas_grid[0], betas_grid[-1], alphas_grid[0], alphas_grid[-1]], cmap="viridis")
ax.set_xlabel("β (inverse temperature)"); ax.set_ylabel("α (learning rate)")
ax.set_title("Mean aversive sounds received")
plt.colorbar(im, ax=ax, label="Sounds (out of 160)")
plt.savefig("figures/task_c.png"); plt.show()
```

```python
**Interpretation:** The fewest aversive sounds occur with high α (fast learning) combined with strongly negative β (deterministic avoidance). Performance degrades when either α is low (slow learning) or β is near zero (random choosing).
```
