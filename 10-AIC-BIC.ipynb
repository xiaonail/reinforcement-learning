```python
## Model Comparison — AIC and BIC

More parameters always reduce NLL, but risk overfitting. AIC and BIC add a complexity penalty:

$$\text{AIC} = 2 \cdot \text{NLL} + 2p \qquad \text{BIC} = 2 \cdot \text{NLL} + p \cdot \ln(n)$$

where $p$ = number of parameters, $n$ = 160 trials. Lower AIC or BIC means a better balance between fit and simplicity.

AIC and BIC can disagree. 

- **AIC** is more focused on prediction. It tends to favour the model that is expected to predict new data better.
- **BIC** applies a stronger penalty for extra parameters. It tends to favour the simpler model, unless the more complex model improves fit by a clear amount.

In practice, this means that AIC often prefers a more flexible model, whereas BIC more often prefers a more parsimonious one.
```

```python
def aic_bic(nll, n_params, n_obs=N_TRIALS):
    return 2*nll + 2*n_params, 2*nll + n_params*np.log(n_obs)

# NLL at fitted parameters for all models
nll_vals = {"M1": [], "M2": [], "M3": []}
for i in range(N_PARTICIPANTS):
    nll_vals["M1"].append(nll_model1(fitted_m1[i],    choices[i], outcomes[i]))
    nll_vals["M2"].append(nll_model2(fitted_m2[i],    choices[i], outcomes[i]))
    nll_vals["M3"].append(nll_model3(fitted_m3[i],    choices[i], outcomes[i]))

n_p = {"M1": 2, "M2": 3, "M3": 3}

print(f"{'Model':<8} {'Σ NLL':<14} {'Σ AIC':<14} {'Σ BIC':<14} {'p'}")
print("-" * 65)
for m in ["M1", "M2", "M3"]:
    s_nll = sum(nll_vals[m])
    s_aic = sum(aic_bic(n, n_p[m])[0] for n in nll_vals[m])
    s_bic = sum(aic_bic(n, n_p[m])[1] for n in nll_vals[m])
    print(f"{m:<8} {s_nll:<14.2f} {s_aic:<14.2f} {s_bic:<14.2f} {n_p[m]}")
```
